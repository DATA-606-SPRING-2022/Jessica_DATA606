```python
import pandas as pd
import numpy as np
import re

from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
% matplotlib inline

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)

stop_words = STOPWORDS.update(["game",'games', "player", 'play'])
```

```python
#Load Data
GameData = pd.read_csv('./FinalSteamData.csv',  on_bad_lines='skip')
GameData.head()
```

```python
CovidData = GameData[GameData['Release Year'].isin([2021, 2022])]
preCovidData = GameData[GameData['Release Year'].isin([2018, 2019])]
```

```python
DescriptionText = ''
for text in GameData['detailed_description.1']:
    if str(text) != 'nan':
        DescriptionText = DescriptionText + text

wordcloud = WordCloud(max_font_size=50, max_words=200, stopwords = stop_words,color_func=lambda *args, **kwargs: "limegreen", collocations = True ).generate(DescriptionText)
```

```python
wordcloud
```

```python
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
